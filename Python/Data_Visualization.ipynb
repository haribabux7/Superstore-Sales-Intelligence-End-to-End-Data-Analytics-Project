## Setup

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df = pd.read_csv('../Dataset/Cleaned_Data.csv', parse_dates=['order_date','ship_date','order_month'])
sns.set_theme(style='whitegrid')
```

## Histograms and boxplots

```python

fig, ax = plt.subplots(1, 2, figsize=(14,5))
sns.histplot(df['sales'], bins=50, kde=True, ax=ax[0]); ax[0].set_title('Sales Distribution')
sns.boxplot(x=df['profit'], ax=ax[1]); ax[1].set_title('Profit Boxplot')
plt.tight_layout()
# Interpretation: Sales and profit contain extreme values; use median, IQR, and outlier flags when operationalizing decisions.
```

## Bar charts and category comparison

```python

cat = df.groupby('category', as_index=False).agg(revenue=('revenue','sum'), profit=('profit','sum'))
fig, ax = plt.subplots(figsize=(9,5))
cat.plot(kind='bar', x='category', y=['revenue','profit'], ax=ax)
ax.set_title('Revenue and Profit by Category')
plt.tight_layout()
# Interpretation: Categories with high revenue but weak profit should be reviewed for discount and shipping policy.
```

## Heatmap

```python

corr = df[['sales','quantity','discount','profit','ship_days','profit_margin']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='RdYlGn', center=0)
plt.title('Correlation Heatmap')
plt.show()
# Interpretation: Discount is the key lever to monitor because it can erode profitability.
```

## Interactive Plotly charts

```python

monthly = df.groupby('order_month', as_index=False).agg(revenue=('revenue','sum'), profit=('profit','sum'))
fig = px.line(monthly, x='order_month', y=['revenue','profit'], markers=True, title='Interactive Monthly Revenue and Profit Trend')
fig.show()

sub = df.groupby('sub_category', as_index=False).agg(revenue=('revenue','sum'), profit=('profit','sum')).sort_values('revenue', ascending=False)
fig = px.bar(sub, x='sub_category', y='revenue', color='profit', title='Sub-Category Revenue Colored by Profit', color_continuous_scale='RdYlGn')
fig.show()

fig = px.scatter(df.sample(min(2500, len(df)), random_state=42), x='discount', y='profit', size='sales', color='category', hover_data=['product_name','region'], title='Discount vs Profit by Category')
fig.show()
```
